**INTERNACOES POR PROCEDIMENTO (ANO E MES DO PROCESSAMENTO)**  - PERIODO JANEIRO DE 2008- JANEIRO DE 2024

```python
#CRIACAO DO DF E LEITURA DOS DADOS CSV
df1_ambulatorio = pd.read_csv('sih_cnv_qirs18000145_162_105_77.csv', sep=';',encoding='latin1',skiprows=4,nrows=2)
df1_ambulatorio
```

```python
#APAGANDO LINHA TOTAL PARA FINS DE ANALISE
df1_ambulatorio.drop(1)
```

```python
# UTILIZAR METODO TRANSPOR PARA TROCAR AS LINHAS POR COLUNAS PARA FACILITAR A ANALISE
df_transposto = df1_ambulatorio.transpose()
df_transposto.head()
```

```python
#MANIPULACAO DO DF PARA FAZER ANALISE EM GRAFICOS
#SELECIONANDO AS COLUNAS QUE REPRESENTAM OS MESES NO DF

meses = df1_ambulatorio.columns[1:-1]

#SOMA DAS QUANTIDADES PARA CADA MES DO PROCEDIMNENTO 

quantidades_por_mes = df1_ambulatorio[meses].sum()

# PLOTANDO UM GRFICO DE LINHAS
plt.figure(figsize=(12, 6))   
quantidades_por_mes.plot(marker='o', linestyle='-')

# PERSONALIZAÇÃO DO GRAFICO
plt.title('Quantidade de Internações por mês: CIRURGIAS COMPLEMENTARES ')
plt.xlabel('Mês')
plt.ylabel('Quantidade')
# Definindo os ticks do eixo x
plt.xticks(range(len(meses)), meses, rotation=45)  # Definindo o espaçamento e a rotação dos ticks

#EXIBIR
plt.tight_layout()
plt.grid(True)
plt.show()
```

**VALOR TOTAL POR PROCEDIMENTO (ANO E MES DO PROCESSAMENTO**) - JANEIRO 2008 -JANEIRO 2024

```python
# Valor total por Procedimento e Ano/mês processamento
# Período:Jan/2008-Jan/2024

df2_ambulatorio = pd.read_csv('sih_cnv_qirs18003645_162_105_77.csv', sep=';',encoding='latin1',skiprows=4,nrows=2)
df2_ambulatorio
```

```python
#APAGANDO LINHA TOTAL
df2_ambulatorio.drop(1)
```

```python
# UTILIZAR METODO TRANSPOR PARA TROCAR AS LINHAS POR COLUNAS PARA FACILITAR A ANALISE
df2_transposto = df2_ambulatorio.set_index('Procedimento').T

#SELECIONAR AS COLUNAS MESES
meses = df2_transposto.index

# PLOTAR GRAFICO
plt.figure(figsize=(10, 6))
plt.plot(meses, df2_transposto.iloc[:, 0], marker='o', linestyle='-')
plt.title('Custo Mensal do Procedimento')
plt.xlabel('Mês')
plt.ylabel('Custo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()

```

**DIAS DE PERMANENCIA POR PROCEDIMENTO(MES E ANO)** - JANEIRO 2008 -JANEIRO 2024

```python
df4_ambulatorio = pd.read_csv('sih_cnv_qirs18010545_162_105_77.csv', sep=';',encoding='latin1',skiprows=4,nrows=2)
df4_ambulatorio
```

```python
#APAGANDO LINHA TOTAL
df4_ambulatorio.drop(1)
```

```python

#REMOVER LINHA TOTAL
df4_ambulatorio = df4_ambulatorio[df4_ambulatorio['Procedimento'] != 'Total']

#SELECIONAR COLUNAS MESES
meses = df4_ambulatorio.columns[1:-1]

# CALCULAR A SOMA
soma_por_mes = df4_ambulatorio[meses].sum()

# EXIBIR GRAFICO

plt.figure(figsize=(10, 6))
soma_por_mes.plot(kind='bar')
plt.title('Quantidade de Dias de Internação por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Dias de Internação')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
